Lo primero a realizar es crear el entorno donde el gato y el raton se van a mover

```python
def crear_laboratorio(fila, columna):
    tablero = [["_" for _ in range(columna)] for _ in range(fila)]
    return tablero

def mostrar_laboratorio(tablero_a_imprimir):
    for fila in tablero_a_imprimir:
        print(" ".join(fila))

f = int(input("Ingrese el numero de filas: "))
c = int(input("Ingrese el numero de columnas: "))


mi_tablero = crear_laboratorio(f, c)
mostrar_laboratorio(mi_tablero)
```
