Lo primero a realizar es crear el entorno donde el gato y el raton se van a mover

```python
def crear_laboratorio(fila, columna):
    tablero = [["_" for _ in range(columna)] for _ in range(fila)]
    return tablero

def mostrar_laboratorio(tablero_a_imprimir):
    for fila in tablero_a_imprimir:
        print(" ".join(fila))

f = int(input("Ingrese el numero de filas: "))
c = int(input("Ingrese el numero de columnas: "))


mi_tablero = crear_laboratorio(f,c)
mostrar_laboratorio(mi_tablero)
```

Ahora debemos colocar los limites del entorno/laboratorio

```python
def pos_validas(fila,columna, pos):
    f,c=pos
    return 0<=f < fila and 0<=c < columna


```

aca vamos a colocar a los personajes

```python
import random
def colocar_personajes(tablero,ubi_gato,ubi_raton):
    tablero[ubi_gato[0]][ubi_gato[1]] = "G"
    tablero[ubi_raton[0]][ubi_raton[1]] = "R"
    return tablero
def ramdomizar():
    gato_pos = (random.randint(0, f - 1), random.randint(0, c - 1))
    raton_pos = (random.randint(0, f - 1), random.randint(0, c - 1))
    while gato_pos == raton_pos:
        raton_pos = (random.randint(0, f - 1), random.randint(0, c - 1))
    return gato_pos, raton_pos


```

Ahora vamos a actualizar el tablero por cada render

```python
def actualizar_laboratorio(fila,columna,ubi_gato,ubi_raton):
    tablero = [["_" for _ in range(columna)] for _ in range(fila)]

    for p, s in [(ubi_gato, "G"), (ubi_raton, "R")]:
        if pos_validas(fila, columna, p): tablero[p[0]][p[1]] = s

    return tablero


```

Llamamos a todas la funciones

```python
mi_tablero = actualizar_laboratorio(f, c, gato_pos, raton_pos)
mostrar_laboratorio(mi_tablero)
```
